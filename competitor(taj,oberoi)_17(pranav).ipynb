```python
# @title Preprocessing Dataset
import pandas as pd
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import matplotlib.pyplot as plt

nltk.download('vader_lexicon')

df = pd.read_csv('pranavpractical.csv')

df_cleaned = df.dropna()

```

```python
# @title Competitor analysis for the data
sid = SentimentIntensityAnalyzer()

def get_sentiment_score(text):
    sentiment_scores = sid.polarity_scores(text)
    return sentiment_scores['compound']

oberoi_reviews = df_cleaned[df_cleaned['product'] == 'oberoi']['text']
taj_reviews = df_cleaned[df_cleaned['product'] == 'taj']['text']

oberoi_sentiments = [get_sentiment_score(review) for review in oberoi_reviews]
taj_sentiments = [get_sentiment_score(review) for review in taj_reviews]

oberoi_avg_sentiment = sum(oberoi_sentiments) / len(oberoi_sentiments)
taj_avg_sentiment = sum(taj_sentiments) / len(taj_sentiments)



```

```python
# @title Visualization
products = ['Oberoi', 'Taj']
avg_sentiments = [oberoi_avg_sentiment, taj_avg_sentiment]

plt.figure(figsize=(10, 6))
plt.bar(products, avg_sentiments, color=['blue', 'green'])
plt.title('Average Sentiment Scores for Different Hotels')
plt.xlabel('Hotel')
plt.ylabel('Average Sentiment Score')
plt.ylim(-1, 1)
plt.grid(axis='y')
plt.show()
```

```python
print("Average Sentiment Scores:")
print("Oberoi Hotel:", oberoi_avg_sentiment)
print("Taj Hotel:", taj_avg_sentiment)
```
