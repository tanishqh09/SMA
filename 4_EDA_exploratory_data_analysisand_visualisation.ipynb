```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
```

```python
df = pd.read_csv("/content/sample_data/california_housing_train.csv")
df.head()
```

```python
# Compute basic statistics for likes, comments, and retweets
stats_likes = df['Likes'].describe()
stats_comments = df['Comments'].describe()
stats_retweets = df['Retweets'].describe()

# Print the statistics
print("Statistics for Likes:")
print(stats_likes)
print("\nStatistics for Comments:")
print(stats_comments)
print("\nStatistics for Retweets:")
print(stats_retweets)
```

```python
# Plot the data
plt.figure(figsize=(10,6))
df['Hashtags'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Count of Posts by Hashtags')
plt.xlabel('Hashtags')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Plot the data as a pie chart
plt.figure(figsize=(8,8))
df['Hashtags'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['skyblue', 'lightgreen', 'lightcoral', 'orange', 'pink'])
plt.title('Percentage of Posts by Hashtags')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

```python
# Combine all post texts
text = ' '.join(df['Post Text'])

# Generate a word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

# Plot the word cloud
plt.figure(figsize=(12,8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.title('Word Cloud of Post Text')
plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
# Convert 'Comments' column to numeric type
df['Comments'] = pd.to_numeric(df['Comments'], errors='coerce')

# Plot the data as a line graph
plt.figure(figsize=(10,6))
df.groupby('Date')['Comments'].sum().plot(color='skyblue')
plt.title('Total Comments Over Time')
plt.xlabel('Date')
plt.ylabel('Total Comments')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Convert Date column to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Create a pivot table
pivot_table = df.pivot_table(index='Date', columns='Hashtags', values='Comments', aggfunc='sum')

# Plot the data as a heat map
plt.figure(figsize=(12,8))
sns.heatmap(pivot_table, cmap='coolwarm', linewidths=0.5, linecolor='black')
plt.title('Comments Heatmap by Date and Hashtags')
plt.xlabel('Hashtags')
plt.ylabel('Date')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Plot the data as a scatter plot
plt.figure(figsize=(10,6))
plt.scatter(df['Comments'], df['Retweets'], color='blue', alpha=0.7)
plt.title('Scatter Plot of Comments vs Retweets')
plt.xlabel('Comments')
plt.ylabel('Retweets')
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Plot the data as a histogram
plt.figure(figsize=(10,6))
plt.hist(df['Comments'], bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram of Comments')
plt.xlabel('Comments')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
